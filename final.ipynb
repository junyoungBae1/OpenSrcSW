```python
import csv
import numpy as np
import matplotlib.pylab as plt

f = open('country_vaccinations.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)
max = 0
min = 999999
user = 0
vaccin = ""
vaccindata = []
maxdate = ""
mindate = ""
userdata = []
i= 0
country = [""]
for row in data:
    if(row[7] == ''):
        row[7] = 0
    else: 
        row[7] = float(row[7])
        if(country[i] != row[0]):
            country.append(row[0])
            i+=1
            vaccindata.append(row[12])
            userdata.append(user)
            user = 0
            
        if max < row[7]:
            max = row[7]
            maxdata = row[2]
        if min > row[7]:
            min = row[7]
            mindata = row[2]
            
        user += float(row[7])
  
        
        
    
        
print("가장 많이 접속 된 날",maxdata)
print("가장 적게 접종 된 날",mindata)


plt.title("vaccine")
plt.pie(userdata,labels = vaccindata,autopct="%.1f%%")
plt.show()
```
